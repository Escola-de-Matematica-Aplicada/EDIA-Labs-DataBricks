```python
%sql
DROP SCHEMA IF EXISTS lodlog_op CASCADE;
DROP SCHEMA IF EXISTS lodlog_dw CASCADE;
```

```python
%sql
-- Lendo o EXPLAIN: Por que a Query é Lenta?
SELECT dc.razao_social, COUNT(*), AVG(fe.atraso_min)
FROM   lodlog_dw.fato_entregas fe
  JOIN lodlog_dw.dim_cliente dc ON fe.sk_cliente = dc.sk_cliente
WHERE  fe. >= '2026-01-01'
GROUP BY dc.razao_social;
```
